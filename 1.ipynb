```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load the MFCC coefficients from the CSV file
csv_file = r"C:\Users\asus\Desktop\DS203\PROJECT\95-MFCC.csv"
mfccs = pd.read_csv(csv_file, header=None).values  # Load without a header if there's none

# Plot the MFCCs
plt.figure(figsize=(10, 6))
plt.imshow(mfccs, aspect='auto', origin='lower', cmap='seismic', interpolation='nearest')
plt.colorbar(label='MFCC Coefficients (Amplitude)')
plt.xlabel('Time Frames')
plt.ylabel('MFCC Coefficients')
plt.title('MFCCs (Cepstral Coefficients) over Time')
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Example data (replace this with your actual MFCC data)
csv_file = r"C:\Users\asus\Desktop\DS203\PROJECT\95-MFCC.csv"
data = pd.read_csv(csv_file, header=None).values  # Load without a header if there's none

# Standardization (Z-score normalization)
data_mean = np.mean(data)
data_std = np.std(data)
standardized_data = (data - data_mean) / data_std

# Visualize
plt.figure(figsize=(12, 8))
plt.imshow(standardized_data, cmap='seismic', aspect='auto', vmin=-3, vmax=3)
plt.colorbar(label='Standardized MFCC Coefficients (Amplitude)')
plt.title("Standardized MFCCs (Cepstral Coefficients) over Time")
plt.xlabel("Time Frames")
plt.ylabel("MFCC Coefficients")
plt.show()

```
