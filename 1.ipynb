```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
# Example data (replace this with your actual MFCC data)
csv_file = "MFCC-files-v2/95-MFCC.csv"
data = pd.read_csv(csv_file, header=None).values  # Load without a header if there's none

# Standardization (Z-score normalization)
data_mean = np.mean(data)
data_std = np.std(data)
standardized_data = (data - data_mean) / data_std

# Visualize
plt.figure(figsize=(12, 8))
plt.imshow(standardized_data, cmap='seismic', aspect='auto', vmin=-3, vmax=3)
plt.colorbar(label='Standardized MFCC Coefficients (Amplitude)')
plt.title("Standardized MFCCs (Cepstral Coefficients) over Time")
plt.xlabel("Time Frames")
plt.ylabel("MFCC Coefficients")
plt.show()

```

```python
plt.axis('off')  # Hides the axes
plt.gca().set_frame_on(False)  # Turns off the frame

# Save the plot with only the graph part
plt.savefig('graph_only.png', bbox_inches='tight', pad_inches=0, transparent=True)
```
